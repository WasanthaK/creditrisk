```python
import requests
import os
from dotenv import load_dotenv

# Load environment variables
load_dotenv()
api_key = os.getenv("EMBEDDING_KEY")
endpoint = os.getenv("EMBEDDING_END_POINT")

# Define the URL for the embeddings endpoint
model_deployment = "text-embedding-ada-002"  # Replace with your actual deployment name if different
api_version = "2024-04-01-preview"

# Construct the full URL
url = f"{endpoint}/openai/deployments/{model_deployment}/embeddings?api-version={api_version}"

# Define the headers
headers = {
    "api-key": api_key,
    "Content-Type": "application/json"
}

# Define the data payload
data_payload = {
    "input": "This is a test input to verify if the endpoint is working."
}

# Perform the API request
try:
    response = requests.post(url, headers=headers, json=data_payload)

    # Check response status code
    if response.status_code == 200:
        print("Success! API endpoint is accessible and returned a valid response.")
        # Print the response content (optional)
        print(response.json())
    else:
        print(f"Failed! Status code: {response.status_code}")
        print(f"Response message: {response.text}")

except requests.exceptions.RequestException as e:
    print(f"An error occurred while trying to connect to the API: {e}")
```
